````python
import urllib.request

def extract_data(url, filename):
  try:
    urllib.request.urlretrieve(url, filename)

  except Exception as e:
    print(e)
````

```
{
    "Version": "2012-10-17",
    "Statement": [
        {
            "Sid": "PublicReadGetObject",
            "Effect": "Allow",
            "Principal": "*",
            "Action": "s3:GetObject",
            "Resource": "arn:aws:s3:::bucket-name/*"
        }
    ]
}
```
